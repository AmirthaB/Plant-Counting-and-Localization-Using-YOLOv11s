```python
%%writefile /content/banana_dataset_new/data/banana_dataset.yaml

path: /content/banana_dataset_new/data

train: train/images
val: valid/images
test: test/images

names:
  0: banana
```

```python
!cat /content/banana_dataset_new/data/banana_dataset.yaml
```

```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")
```

```python
results = model.train(
    data="/content/banana_dataset_new/data/banana_dataset.yaml",
    epochs=250,
    imgsz=960,
    batch=4,
    patience=50,
    workers=2,
    cache=True,
    device=0,
    project="banana_segmentation",
    name="yolo11s"
)
```

```python
metrics=model.val()
print(metrics)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/banana_segmentation/yolo11s/weights/best.pt")
```

```python
results = model.predict(
    source="/content/banana12_plant.jpg",
    conf=0.10,
    imgsz=960,
    save=True
)
```

```python
count = len(results[0].boxes)
print("Number of banana plants detected:", count)
```
